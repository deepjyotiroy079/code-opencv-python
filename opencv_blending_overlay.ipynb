# Overlay (without blending)

This is done when we have two images of different sizes.

```python
small_img = cv2.imread('Images/no_copy.jpg')
# converting the colours
small_img = cv2.cvtColor(small_img, cv2.COLOR_BGR2RGB)

large_img = cv2.imread('Images/building.jpg')
# converting the colours
large_img = cv2.cvtColor(large_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(large_img)
```

```python
plt.imshow(small_img)
```

We need to give a starting point of the second image and the ending point of the second image. Usually represented by x_offset and y_offset. And for end we have x_end and y_end

```python
# second image should start at (0, 0)
x_offset = 0
y_offset = 0
```

```python
# x_end and y_end are basically the offset + the height and width of the second image
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

Replace chunks of the larger image with the smaller image using the starting point as the offset and the ending point as the end

```python
# we are just replacing the chunk of pixels in the larger image with the smaller image.
large_img[y_offset:y_end, x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
