```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/image1.png')
```

```python
cv2.imshow('Image', img)
cv2.waitKey(0)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# MATPLOTLIB --> RED, GREEN, BLUE
# OPENCV --> BLUE, GREEN, RED

# CONVERTING BGR TO RGB
converted_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(converted_image)
```

```python
# converting to gray scal
img_gray = cv2.imread('Images/image1.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
# 500 mentions the height
# 200 is the width
# -------------------------------
new_img = cv2.resize(converted_image, (500, 200))
```

```python
plt.imshow(new_img)
```
