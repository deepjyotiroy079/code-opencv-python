```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('Images/image1.png')
```

```python
numpy_pic = np.asarray(pic)
```

```python
numpy_pic.shape
```

```python
pic_red = numpy_pic.copy()
pic_red[:, :, 1] = 0
pic_red[:, :, 2] = 0
```

```python
plt.imshow(pic_red)
```

```python
pic_blue = numpy_pic.copy()
pic_blue[:, :, 0] = 0
pic_blue[:, :, 1] = 0

plt.imshow(pic_blue)
```

```python
pic_green = numpy_pic.copy()
pic_green[:, :, 0] = 0
pic_green[:, :, 2] = 0

plt.imshow(pic_green)
```

```python
plt.figure()
f, arr = plt.subplots(1, 3)
arr[0].imshow(pic_red)
arr[1].imshow(pic_green)
arr[2].imshow(pic_blue)
f.tight_layout()
```
